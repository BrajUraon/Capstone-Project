## Encoding

```python
# OneHotEncoder:
# Columns:- fuel, transmission, drive, color

# OrdinalEncoder:
# Columns:- cylinders,condition

# FrequencyEncoding:
# Columns:- manufacturer, model, type, state, title_status

# Rest of the columns:- yet to be determined
```

```python
df2= df.copy()
```

```python
# Frequency Encoding

# Manufacturer

dict1 = df2['manufacturer'].value_counts().to_dict()

for key,values in dict1.items():
    for i in df2['manufacturer']:
        if i == key:
            df2.loc[df2['manufacturer']==key,'manufacturer'] = values
```

```python
# Model 

dict1 = df2['model'].value_counts().to_dict()

for key,values in dict1.items():
    for i in df2['model']:
        if i == key:
            df2.loc[df2['model']==key,'model'] = values
```

```python
# state 

dict3 = df2['state'].value_counts().to_dict()

for key,values in dict3.items():
    for i in df2['state']:
        if i == key:
            df2.loc[df2['state']==key,'state'] = values
```

```python
# Type 

dict3 = df2['type'].value_counts().to_dict()

for key,values in dict3.items():
    for i in df2['type']:
        if i == key:
            df2.loc[df2['type']==key,'type'] = values
```

```python
# title_status 

dict4 = df2['title_status'].value_counts().to_dict()

for key,values in dict4.items():
    for i in df2['title_status']:
        if i == key:
            df2.loc[df2['title_status']==key,'title_status'] = values
```
